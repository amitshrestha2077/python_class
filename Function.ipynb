```python
def add(a,b):
    return a+b
def sub(a,b):
    return a-b
def mul(a,b):
    return a*b
def div(a,b):
    return a/b

def call(a,b,o):
    if o == "+":
        print(add(a,b))
    elif o == "-":
        print(sub(a,b))
    elif o == "*":
        print(mul(a,b))
    elif o == "/":
        print(div(a,b))
    else:
        print("Invalid opration")
        

a = int(input("Enter a = "))
b = int(input("Enter a = "))
o = input("ENter + or - or * or /")       
call(a,b,o)
    
```

```python
#  filter
# list(filter(lambda <argument>:<expression>,list))
```

```python
l = [32,53,43,31,42,31,4,42,56,21,32,1]
list(filter(lambda x:x%2==0,l))
```

```python
l = ["Apple","Cat","Dog","Fish","Goat","ANT"]    
list(filter(lambda x:x.isupper(),l))      #capital later
```

```python
l = ["Apple","cat","Dog","Fish","Goat","ANT"]    
list(filter(lambda x:x.islower(),l))      #smaller later
```

```python
# map function
# list(map(lambda <argument>:<expression>,list))
```

```python
salary = [1000000,2000000000,50000000,30000000000,4320000]
list(map(lambda x:int(x*1.1),salary))
```

```python
l = ['Apple','cat','dog','Fish','Goat','ant']    
list(map(lambda x:x.lower(),l))
```
